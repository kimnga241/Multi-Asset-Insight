```python
# 1. Day Parse 
vnindex ['Ngày'] = pd.to_datetime(vnindex['Ngày'], format='%Y-%m-%d')
gold_pr ['Ngày'] = pd.to_datetime(gold_pr ['Ngày'], format='%m/%d/%Y')
usdvnd  ['Ngày'] = pd.to_datetime(usdvnd  ['Ngày'], format='%Y-%m-%d')

# 2. Change ColumnName
vn2 = (
    vnindex
    .rename(columns={'Lần cuối':'vnindex'})
    .set_index('Ngày')
)
gp2 = (
    gold_pr
    .rename(columns={'Bán ra':'Giá Vàng'})
    .set_index('Ngày')
)
usd2 = (
    usdvnd
    .rename(columns={'Lần cuối':'usdvnd'})
    .set_index('Ngày')
)

# 3. Concat 
df_all = pd.concat([vn2, gp2, usd2], axis=1, join='inner').reset_index()
df_all = df_all.drop(columns=['% Thay đổi', 'Mua vào'])


# 4. Filter for Month-End
df_all['Ngày'] = pd.to_datetime(df_all['Ngày'])

ym = df_all['Ngày'].dt.to_period('M')
idx = df_all.groupby(ym)['Ngày'].idxmax()
df_max_per_month = df_all.loc[idx].reset_index(drop=True)

df_max_per_month.to_csv('finnal.csv', index=False)


```
